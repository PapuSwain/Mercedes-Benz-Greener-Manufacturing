```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
train=pd.read_csv("train1.csv")
test=pd.read_csv("test1.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
train.drop(columns="ID",inplace=True)
test.drop(columns="ID",inplace=True)
```

```python
train.shape,test.shape
```

## Remove columns having zero variance

```python
zero_var_col=train.var()[train.var()==0].index.values
zero_var_col
```

```python
train.drop(zero_var_col,axis=1,inplace=True)
test.drop(zero_var_col,axis=1,inplace=True)
```

```python
train.shape,test.shape
```

## Checking of null values in Train and Test dataset

```python
train.isna().sum()
```

```python
np.sum(train.isna().sum())
```

```python
test.isna().sum().sum()
```

## Checking of  unique values in Train and Test dataset

```python
train.describe(include="all").T
```

```python
train.nunique()
```

```python
test.nunique()
```

```python
for i in train:
    print("\nThe unique values in column ",i)
    print(train[i].unique())
```

```python
for i in test:
    print("\nThe unique values in column ",i)
    print(test[i].unique())
```

```python
train.head()
```

## Apply Label Encoder to the columns having object datatype

```python
train.describe(include="object")
```

```python
object_col=train.describe(include="object").columns.values
object_col
```

```python
test.describe(include="object").columns.values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
for i in object_col:
    label_encoder.fit(train[i].append(test[i]).values)
    train[i]=label_encoder.transform(train[i])
    test[i]=label_encoder.transform(test[i])
```

```python
train.head()
```

## Building ML model

```python
from sklearn.model_selection import train_test_split
```

```python
x=train.drop(columns="y")
y=train[["y"]]
```

```python
x.head()
```

```python
y.head()
```

```python
x.shape,y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=25)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

## Perform dimensionality reduction(PCA-Principal Component Analysis)

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.95)
pca.fit(x_train)
```

```python
pca.n_components_
```

```python
pca.explained_variance_
```

```python
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.transform(x_test)
```

```python
x_train_pca
```

```python
x_test_pca
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model=LinearRegression()
linear_model.fit(x_train_pca,y_train)
```

```python
pred=linear_model.predict(x_test_pca)
pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_pred=pred,y_true=y_test)
```

```python
mean_squared_error(y_pred=pred,y_true=y_test,squared=False)
```

```python
r2_score(y_pred=pred,y_true=y_test)
```

## Linear Regression using XGBoost

```python
from xgboost import XGBRegressor
```

```python
xgb_model=XGBRegressor(objective="reg:linear",learning_rate=0.1)
xgb_model.fit(x_train_pca,y_train)
```

```python
xgb_pred=xgb_model.predict(x_test_pca)
xgb_pred
```

```python
mean_squared_error(y_pred=xgb_pred,y_true=y_test,squared=False)
```

```python
r2_score(y_pred=xgb_pred,y_true=y_test)
```

## Random forest using XGBoost

```python
from xgboost import XGBClassifier
```

```python
xgb_rf_model=XGBClassifier()
xgb_rf_model.fit(x_train_pca,y_train)
```

```python
xgb_rf_pred=xgb_rf_model.predict(x_test_pca)
xgb_rf_pred
```

```python
mean_squared_error(y_pred=xgb_rf_pred,y_true=y_test,squared=False)
```
